```python
import pandas as pd
df=pd.read_csv("/content/movies_with_clusters.csv")
df
```

```python
features=df.filter(like="genre_").copy()
features["rating"]=df["rating"]
```

```python
from sklearn.preprocessing import StandardScaler
X=StandardScaler().fit_transform(features)
```

```python
from sklearn.cluster import AgglomerativeClustering
df["cluster"]=AgglomerativeClustering(n_clusters=5).fit_predict(features)
```

```python
def recommend(title,top_n=3):
  cluster_id=df.loc[df['title']==title,'cluster'].iloc[0]
  sim=df.query("cluster==@cluster_id and title!=@title").nlargest(top_n,'rating')
  return "\n".join(f"{r.title} (Rating: {r.rating}) - Do you want to watch also?" for _, r in sim.iterrows())
```

```python
user_input=input("Enter the movie name:")
print(f"\n Recommendation similar to{user_input}\:n")
print(recommend(user_input))
```

```python
import pandas as pd

from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data =[

    {'Milk':1,'Bread':1,'Sugar':0,'Teapowder':1},
    {'Milk':0,'Bread':1,'Sugar':1,'Teapowder':1},
    {'Milk':0,'Bread':1,'Sugar':0,'Teapowder':1},
    {'Milk':1,'Bread':1,'Sugar':0,'Teapowder':1},
    {'Milk':0,'Bread':1,'Sugar':0,'Teapowder':1}
]
df=pd.DataFrame(data)
df
```

```python
frequent_itemsets=apriori(df,min_support=0.6,use_colnames=True)
```

```python
rules=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.8)
```

```python
print("frequently picked items ",frequent_itemsets)
```
